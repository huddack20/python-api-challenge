```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Google developer API key
from api_keys import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
output_data_path = "../output_data/cities.csv"

# Read the mouse data and the study results
cities_df = pd.read_csv(output_data_path)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations = cities_df[["Lat", "Lng"]]

# Fill NaN values and convert to float
cities_df.fillna(0)
rating = cities_df["Humidity"].astype(float)

# Plot Heatmap
center_coordinates = (0.00, 0.00)
fig = gmaps.figure(center=center_coordinates, zoom_level=1)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=250,
                                 point_radius=4.0)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig

```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Narrow down the DataFrame to find your ideal weather condition.
filtered_cities_df = cities_df

# Drop any rows that don't contain all three conditions. Want to be sure the weather is ideal.

# A max temperature lower than 80 degrees but higher than 70.
filtered_cities_df = filtered_cities_df.loc[(filtered_cities_df["Max Temp"] < 80) & (filtered_cities_df["Max Temp"] > 70)]

# Wind speed less than 10 mph.
filtered_cities_df = filtered_cities_df.loc[filtered_cities_df["Wind Speed"] < 10]

# Zero cloudiness.
filtered_cities_df = filtered_cities_df.loc[filtered_cities_df["Cloudiness"] == 0]

# Drop any rows with null values
filtered_cities_df = filtered_cities_df.dropna()

filtered_cities_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = filtered_cities_df

# params dictionary to update each iteration
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key
}

for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {row['City']}.")
    response = requests.get(base_url, params=params).json()
    #print(json.dumps(response, indent=4, sort_keys=True))
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel is {results[0]['name']}.")
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field of result... skipping.")
        
    print("------------")
```

```python
# Assign the marker layer to a variable
locations = hotel_df[["Lat", "Lng"]]
markers = gmaps.marker_layer(locations)
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]

# Create hotel symbol layer
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=hotel_info
)
# Add the layer to the map
fig.add_layer(markers)
fig.add_layer(hotel_layer)
fig
```
